```python
# gropuby 1, 칼럼이름을 바꿔줘야 할 때 코드가 길어질 수 있다.
df.groupby('continent').count()[['country']].rename(columns = {'country':'country_count'})
```

```python
scientists['age_days_dt'] = scientists['died_dt'] - scientists['born_dt']
scientists.head()
```

```python
scientists_dropped = scientists.drop(['Died', 'Age'], axis = 1)
scientists_dropped
```

```python
data = pd.read_csv('data/@preprocessing_data_member01.csv', encoding = 'utf-8')

# 최종주문요일을 Categorical data로 수정합니다.
data['최종주문요일'] = pd.Categorical(data['최종주문요일'],
                                ['Monday','Tuesday','Wednesday','Thursday','Friday' ,'Saturday','Sunday'])
data.head()
```

```python
data.groupby(['최종주문요일', '성별'])['총구매횟수'].sum()
```

```python
pd.pivot_table(data = data, index = '최종주문요일', values = '총구매횟수')
```
