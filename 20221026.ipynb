```python
import pandas as pd
uselog = pd.read_csv('data/스포츠센터데이터/use_log.csv')
customer_join = pd.read_csv('data/스포츠센터데이터/customer_join.csv')
uselog['usedate'] = pd.to_datetime(uselog['usedate'])
uselog['연월'] = uselog['usedate'].dt.strftime('%Y%m')
uselog_months = uselog.groupby(['customer_id','연월'], as_index = False).agg(count = ('log_id', 'count'))
uselog_months.head(3)
```

```python
uselog_pivot = pd.pivot_table(data = uselog_months, 
                              index = 'customer_id',
                              columns = '연월',
                              values = 'count')
uselog_pivot.head()
```

```python
total_column_count = len(uselog_pivot.columns)
set_column_count = 2
result_df = pd.DataFrame()

for i in range(total_column_count-set_column_count+1):
    selected_df = uselog_pivot.iloc[:, i:i+set_column_count]
    selected_df.loc[:,'연월'] = selected_df.columns[-1]
    selected_df.columns = ['before_month_count', 'current_month_count','연월']
    result_df = pd.concat([result_df, selected_df])
    
uselog = result_df.reset_index()
```

```python
from dateutil.relativedelta import relativedelta

exit_customer = customer_join[customer_join.copy().loc[:,'is_deleted'] == 1]
exit_customer.loc[:,'exit_date'] = None
exit_customer.loc[:,'end_date'] = pd.to_datetime(exit_customer.copy().loc[:,'end_date'])
```

```python
exit_customer['exit_date'] = exit_customer['end_date'].apply(lambda x: x - relativedelta(months = 1))
```

```python
exit_dates = []
for date in exit_customer.loc[:,'end_date']:
    diff_date = date - relativedelta(months = 1)
    exit_dates.append(diff_date)
exit_customer.loc[:,'exit_date'] = exit_dates
```

```python
exit_customer = exit_customer.copy()
exit_customer.loc[:,'연월'] = exit_customer.loc[:,'exit_date'].dt.strftime('%Y%m')
exit_customer.head()
```

```python
exit_uselog = exit_customer.merge(uselog, on = ['customer_id','연월'])
exit_uselog.head()
```

```python
conti_customer = customer_join[customer_join.copy().loc[:,'is_deleted'] == 0]
conti_uselog = conti_customer.merge(uselog,
                    on = ['customer_id'])
```

```python
conti_uselog = conti_uselog.sample(frac = 1) # 랜덤섞기
conti_uselog = conti_uselog.drop_duplicates(subset = ['customer_id']) # 중복제거
```

```python
predict_data = pd.concat([exit_uselog,conti_uselog], ignore_index = True)
```

```python
predict_data = predict_data.dropna(subset = ['before_month_count'])
```

```python
target_col = [ 'gender', 'is_deleted', 'class_name', 'membership_period',
              'campaign_name','routine_flg', 'before_month_count']
```

```python
dataset = predict_data[target_col]
dataset = pd.get_dummies(dataset)
drop_dataset = dataset.drop(['gender_M','campaign_name_2_일반',
                             'class_name_2_야간'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# 탈퇴유저, 지속유저의 데이터의 수를 일치시킵니다.
exit = drop_dataset[drop_dataset['is_deleted'] == 1]
conti = drop_dataset[drop_dataset['is_deleted'] == 0].sample(len(exit))
concat_data = pd.concat([exit, conti])

X = concat_data.drop(['is_deleted'], axis = 1)
Y = concat_data['is_deleted'].ravel()

X_train, X_test, Y_train, Y_test = train_test_split(
    X,Y, test_size = 0.3, random_state = 0)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
train_accuracy = model.score(X_train, Y_train)
test_accuracy = model.score(X_test, Y_test)
print(f'훈련 정확도 : {train_accuracy}')
print(f'테스트 정확도 : {test_accuracy}')
```

```python
model_none_overfitting = DecisionTreeClassifier(max_depth = 3)
model_none_overfitting.fit(X_train, Y_train)
```

```python
train_accuracy = model_none_overfitting.score(X_train, Y_train)
test_accuracy = model_none_overfitting.score(X_test, Y_test)
print(f'훈련 정확도 : {train_accuracy}')
print(f'테스트 정확도 : {test_accuracy}')
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(max_depth = 5,
                                n_estimators = 100)
model_rf.fit(X_train,Y_train)

pred_model_rf = model_rf.predict(X_train)

rf_train_score = model_rf.score(X_train, Y_train)
rf_test_score = model_rf.score(X_test, Y_test)

print(f'랜덤포레스트 훈련 정확도는 {round(rf_train_score,2)} 입니다.')
print(f'랜덤포레스트 테스트 정확도는 {round(rf_test_score,2)} 입니다.')
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbrt = GradientBoostingClassifier(max_depth = 4, learning_rate = 0.05)
gbrt.fit(X_train, Y_train)

gbrt_train_score = gbrt.score(X_train, Y_train)
gbrt_test_score = gbrt.score(X_test, Y_test)

print(f'그레디언트부스팅 훈련 정확도는 {round(gbrt_train_score,2)} 입니다.')
print(f'그레디언트부스팅 테스트 정확도는 {round(gbrt_test_score,2)} 입니다.')
```
