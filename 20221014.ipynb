```python
data = pd.read_csv('data/@preprocessing_data_member01.csv', encoding = 'utf-8')

#최종주문요일을 Ctegorical data로 수정합니다.
data['최종주문요일'] = pd.Categorical(data['최종주문요일'],
['Monday','Tuesday','Wednesday','Thursday','Friday' ,'Saturday','Sunday'])
data
```

```python
data.groupby('최종주문요일').mean()
```

```python
data.groupby('최종주문요일')['총구매횟수'].mean()
```

```python
data.groupby(['최종주문요일','성별'])['총구매횟수'].mean()
```

```python
data.groupby(['최종주문요일','성별'])['총구매횟수'].sum()
```

```python
pd.pivot_table(data = data, index = '최종주문요일', values = '총구매횟수', aggfunc ='sum')
```
